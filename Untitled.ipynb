```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = '../../data/IR-MNIST/'
# directory to save state_dict and loss.npy
save_dir = data_dir + 'save/'
```

```python
train_loss = np.load(save_dir + 'train_loss_list.npy')
validation_loss = np.load(save_dir + 'validation_loss_list.npy')
```

```python
plt.title('Inpainter')
plt.plot(range(len(train_loss[0])), train_loss[0])
plt.plot(range(len(validation_loss[0])), validation_loss[0])
```

```python
plt.title('Detector')
plt.plot(range(len(train_loss[1])), train_loss[1])
plt.plot(range(len(validation_loss[1])), validation_loss[1])
```
